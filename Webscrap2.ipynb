```python
# Importing libraries
import requests
import csv
from bs4 import BeautifulSoup
from selenium  import webdriver
import pandas as pd
import numpy as np
# code for web scraping
driver = webdriver.Chrome("/home/shantam/Downloads/chromedriver")
driver.get('https://www.prokabaddi.com/stats/49-102-total-points-statistics')
element = driver.find_element_by_id('load_more')
for i in range(17):
    element.click()
soup = BeautifulSoup(driver.page_source)
for item in soup.find_all("div",class_="sipk-lb-playerName"):
    [elem.extract() for elem in soup.find("div")]
    f = open("parsed.txt", "a")
    f.write(item.text)
    f.close()    
playerDetails = pd.read_csv('parsed.txt',encoding='ISO-8859-1',sep='\t',names=['playerName'])
print(playerDetails)
    


import os
if os.path.exists("parsed.txt"):
    os.remove("parsed.txt")
  
```

```python
# Adding a column named as PlayerName in the dataframe
df=pd.DataFrame(data=[])
df['PlayerName'] = playerDetails[playerDetails.columns[0]]
df
```
