```python
# successful_raid -Season1
from selenium import webdriver                  
from bs4 import BeautifulSoup
import time
import pandas as pd
import requests
url="https://www.prokabaddi.com/stats/1-13-successful-raids-statistics"
driver=webdriver.Chrome("/home/shantam/Downloads/chromedriver")
data = driver.get(url)
time.sleep(20)
'''element=driver.find_element_by_xpath('//*[@id="load_more"]')
for a in range(7):
    time.sleep(5)
    element.click()'''
html=driver.page_source
soup=BeautifulSoup(html, 'html.parser')
name=[]
matches_played=[]
successful_raid=[]
element1=soup.find_all('div',{"class":"lb-detailItem-wrap"})
for a in element1:
    result=a.find('div',{"class":"sipk-lb-playerName"}).text
    result=result.strip(' \t\n')
    name.append(result)
for b in element1:
    result2=b.find('div',{"class":"sipk-lb-detailBlock sipk-lb-matchedPlayed"}).text
    result2=result2.strip(' \t\n')
    matches_played.append(result2)
for c in element1:
    result3=c.find('div',{"class":"sipk-lb-detailBlock sipk-lb-raidPoints"}).text
    result3=result3.strip('\n')
    successful_raid.append(result3)
i=0
df=pd.DataFrame({'Name':name,'Matches_Played':matches_played,'Successful_raid':successful_raid})
df.to_csv('Season_1_Team_successful_raid.csv',sep=',', encoding='utf-8')
```
